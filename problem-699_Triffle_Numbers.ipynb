First we need to calculate \sigma(n)=the sum of all divisors of the positive integer n.
We calculate all \sigma(n) for 0<n<N+1 and we compare (\sigma(n)/n) in it's lowest form a/b and if b=3^k then
We sum all of n that verifies the condition and that's the number we are looking for

```python
def sum_divisors(n):
    sum = 0
    divisor=1
    if(n==0):
        return sum
    while divisor<=n/2:
        if n%divisor==0:
            sum+=divisor
            divisor+=1
        else:
            divisor+=1
    sum+=n
  # Return the sum of all divisors of n, not including n
    return sum
print("sum of all divisors is: ",sum_divisors(99))

# The following function calculates the sum of all divisors of a number including itself 
#  def sum_divisors(number):
#      if number<=0:
#           sum = 0
#            print("This number is not a positive integer")
#        else:
#           sum = 1
#           for i in range(2,int(number+1)):
#                if number%i==0:
#                    sum +=i
#        return sum    
#print("sum of all divisors is: ",sum_divisors(12))
```

```python
# A function that calculates divisors of a number and put them into a list
def divisors(number):
    divisors_list = []
    for i in range(1,number+1):
            if number%i==0:
                divisors_list.append(i)
    return divisors_list
print("all divisors are:", divisors(156))
```

```python
# Calculate the lowest form of a fraction
def lowest_fraction_form(numerator,denominator):
    result_dict={}
    numerator_divisors = divisors(numerator)
    numerator_divisors.reverse()
    #print(numerator_divisors)
    denominator_divisors = divisors(denominator)
    denominator_divisors.reverse()
    #print(denominator_divisors)
    for items in denominator_divisors:
        #print("item is:",items)
        for otheritems in numerator_divisors:
            #print("otheritem is:",otheritems)
            if items == otheritems:
                numerator/=items
                denominator/= otheritems
                result_dict = {"lowest_numerator":numerator,"lowest_denominator":denominator}
                break;
        if result_dict!={}:
            break
    return result_dict
print("lowest_fraction_form:",lowest_fraction_form(156,99))
```

```python
def three_power_k(denominator):
    for k in range(2,denominator):
        if pow(3,k) == denominator:
            return True
    return False

print(three_power_k(81))
```

```python
def triffle(number):
    sum = 0
    if number<=0:
        print("This number is not a positive integer")
    else:
        for n in range(2,number+1):
            #print("n=",n)
            divisors=sum_divisors(n)
            #print("sum of n's divisors=",divisors)
            sigma_n=lowest_fraction_form(divisors,n)
            #print("sigma_n=",sigma_n)
            denom=int(sigma_n.get("lowest_denominator"))
            if three_power_k(denom):
                sum+=n
                #print("lastsum=",sum)
    return sum
print("sum of n's=",triffle(100))
#test
```
